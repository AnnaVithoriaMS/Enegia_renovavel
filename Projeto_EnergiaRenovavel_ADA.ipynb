 # Produção de Energias Renováveis


Energias renováveis são fontes naturais de energia capaz
  de se renovar, diferentemente de  energias de fonte não renovavel, como os combustíveis fósseis.  Exemplos de energia renovável incluem solar, eólica, hidrelétrica, geotérmica e biomassa.
    Assim as fontes são abundantes e inesgotáveis, alem de serem limpas, já que produzem menos ou nenhum gás de efeito estufa, no entanto é importante resaltar que a produção é afeta por codições naturais, alem de dificuldades para armazenamento e distribuição.

```python
def mapa_energia(ener):
    aggs = ["sum","avg","median","mode","min","max"]

    agg = []
    agg_func = []
    for i in range(0, len(aggs)):
        agg = dict(
            args=['transforms[0].aggregations[0].func', aggs[i]],
            label=aggs[i],
            method='restyle'
        )
        agg_func.append(agg)

    data = [dict(
      type = 'choropleth',
      locationmode = 'country names',
      locations = df_ener['Pais'],
      z = df_ener[ener],
      autocolorscale = False,
      colorscale = 'Portland',
      reversescale = True,
      transforms = [dict(
        type = 'aggregate',
        groups = df_ener['Pais'],
        aggregations = [dict(
            target = 'z', func = 'sum', enabled = True)
        ]
      )]
    )]

    layout = dict(
      title = f'Energia {ener} nos países',
      xaxis = dict(title = 'Subject'),
      yaxis = dict(title = 'Score', range = [0,22]),
      height = 600,
      width = 900,
      updatemenus = [dict(
            x = 0.85,
            y = 1.15,
            xref = 'paper',
            yref = 'paper',
            yanchor = 'top',
            active = 1,
            showactive = False,
            buttons = agg_func
      )]
    )

    fig_dict = dict(data=data, layout=layout)

    pio.show(fig_dict, validate=False)
```

```python
mapa_energia('Eolica')
```

```python
mapa_energia('Hidroeletrica')
```

## Brasil

    Durante o periodo que vai de 2000 a 2021 a energia proveniente de hidroeletricas em relação a produção total de energia renovavel no pais foi de 0.88.  No ano de 2002, o governo brasileiro criou o Programa de Incentivo às Fontes Alternativas de Energia Elétrica (Proinfa), cujo objetivo era ampliar as matrizes que geram menos impactos ambientais, dentre elas a eólica, a solar, a geotérmica e outras, nesse período houve um crescimento da produção de energia eólica, entre 2010 e 2020 a produção de energia eolica passou de 2.18 TWh para 57.05 TWh, motivada pela produção de energia eólica nos estados da Região Nordeste do país, o crescimento continua entre 2020 e 2021 houve um aumento de 26.64% na produção.
    
    Já a energia solar teve produção mais significativa a partir de 2015, e foi crescento conforme os anos, existido um salto significativo apos 2018, quadruplicando a produção entre 2018 e 2021, passando de 3.46 TWh para 12.48 Twh. No período também é perceptível o aumento da produção por outras fontes como biomassa;
